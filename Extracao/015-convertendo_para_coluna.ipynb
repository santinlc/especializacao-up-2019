```python
import pandas as pd
import glob
import time
```

```python
import configparser

config = configparser.RawConfigParser()
config.read('parametros.config')

PARAMETROS = dict(config.items('Padronizacao_conversao'))
PARAMETROS_GERAIS = dict(config.items('Geral'))
```

```python
PASTA_DE_DADOS = PARAMETROS_GERAIS['pasta_raiz_dados']+PARAMETROS['caminho_arquivo_convertido']
PASTA_SAIDA = PARAMETROS_GERAIS['pasta_raiz_dados']+"__"+PARAMETROS['caminho_arquivo_convertido']
```

```python
lista_arquivos = [i for i in glob.glob(PASTA_DE_DADOS+"/*")]
```

```python
for arquivo in lista_arquivos:
    indice = arquivo.rindex('/')
    nome_arquivo = arquivo[indice+1:]
    df = pd.read_csv(arquivo,dtype='str')
    lista_contas = ['conta_'+conta for conta in df['CD_CONTA_CONTABIL'].unique().tolist()]
    lista_operadoras = df['REG_ANS'].unique().tolist()
    lista_registros = []
    for _idx,reg_ans in enumerate(lista_operadoras):
        print(time.ctime(),reg_ans,nome_arquivo)
        dict_registro = dict(zip(lista_contas,[0]*len(lista_contas)))
        dict_registro['REG_ANS'] = reg_ans
        df_filtrado = df[df['REG_ANS'] == reg_ans]
        for conta in df_filtrado['CD_CONTA_CONTABIL'].unique().tolist():
            df_conta = df_filtrado[df_filtrado['CD_CONTA_CONTABIL']==conta]
            dict_registro["conta_"+conta] = df_conta['VL_SALDO_FINAL'].sum()
        dict_registro['ANO'] = df_conta["ANO"].sum()
        dict_registro['SEMESTRE'] = df_conta["SEMESTRE"].sum()
        lista_registros.append(dict_registro)       
        print("Finalizado ",reg_ans,"(",(_idx+1),"-",len(lista_operadoras),")")
    print(PASTA_SAIDA+nome_arquivo)
    pd.DataFrame(lista_registros,dtype='str').to_csv(PASTA_SAIDA+nome_arquivo,index=False)
```
