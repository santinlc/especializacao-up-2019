```python
import pandas as pd
```

```python
df_demonstracoes_plano_contas = pd.read_csv('../entrada/dados_demonstracoes_sumarizado.csv',encoding="ISO-8859-1")
```

```python
df_idss = pd.read_csv("../entrada/idss_alto_nivel.csv",encoding="ISO-8859-1")
```

```python
df_demonstracoes = df_demonstracoes_plano_contas[df_demonstracoes_plano_contas["REG_ANS"].isin(df_idss['Registro_ANS'].tolist())]
```

```python
df_demonstracoes["REG_ANS"].head(2)
```

```python
 df_demonstracoes = df_demonstracoes.groupby(['REG_ANS','ANO','SEMESTRE'],as_index=False).sum()
```

```python
df_demonstracoes = df_demonstracoes[[
    'REG_ANS', 'ANO', 'SEMESTRE','PEONA_SUS',
       'PEONA_OUTROS', 'PESL_OUTROS', 'PESL_SUS', 'REMISSAO', 'PPCNG',
       'PROVISOES_TECNICAS', 'DESPESA_ADM', 'DESPESA_COM', 'VARIACAO_PEONA',
       'EVENTOS_INDENIZAVEIS_POS', 'EVENTOS_INDENIZAVEIS_PRE', 'DESPESA',
       'CORRESP_CEDIDA_POS', 'CORRESP_CEDIDA_PRE', 'CONTRAPRESTACAO_POS',
       'CONTRAPRESTACAO_PRE', 'RECEITA', 'PL', 'PASSIVO_NAO_CIRCULANTE',
       'PASSIVO_CIRCULANTE', 'PASSIVO', 'CREDITO_OPER_PS', 'ATIVO_CIRCULANTE',
       'ATIVO']]
```

```python
df_demonstracoes.to_csv("../entrada/dados_demonstracoes_sumarizado.csv",encoding="ISO-8859-1",index=False)
```

```python
df_demonstracoes = df_demonstracoes.merge(df_idss,left_on=['REG_ANS','ANO'],right_on=['Registro_ANS','ano'])
```

```python
df_demonstracoes['RESULTADO_LIQUIDO'] = df_demonstracoes['RECEITA']- df_demonstracoes['DESPESA']
df_demonstracoes["ROE"] = df_demonstracoes['RESULTADO_LIQUIDO']/df_demonstracoes['PL']
df_demonstracoes['MLL'] = df_demonstracoes['RESULTADO_LIQUIDO']/df_demonstracoes['RECEITA']
df_demonstracoes['LC'] = df_demonstracoes['ATIVO_CIRCULANTE']/df_demonstracoes['PASSIVO_CIRCULANTE']

```
